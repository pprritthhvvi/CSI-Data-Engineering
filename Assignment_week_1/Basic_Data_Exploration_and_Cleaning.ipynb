```python
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

for col in numerical_columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_columns = df.select_dtypes(include=["object"]).columns

for col in categorical_columns:
    df[col] = df[col].fillna("Unknown")
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df["initial_price"].dtype)
print(df["final_price"].dtype)
```

```python
df[['initial_price', 'final_price']].head()
```

```python
df['final_price'] = df['final_price'].astype(str)

df['final_price'] = df['final_price'].str.replace('₹', '', regex=False)
df['final_price'] = df['final_price'].str.replace(',', '', regex=False)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
```

```python
df['final_price'].head()
```

```python
df['discount_amount'] = df['initial_price'] - df['final_price']
```

```python
df['discount_amount'] = df['discount']
```

```python
df['discount_amount'] = df['initial_price'] * (df['discount'] / 100)
```

```python
df["quantity"] = 1

df["total_amount"] = df["initial_price"] * df["quantity"]

df["discount_amount"] = (
    df["initial_price"] * df["discount"] / 100
)
```

```python
print("\nDerived Columns")
print(
    df[
        [
            "initial_price",
            "quantity",
            "total_amount",
            "discount_amount"
        ]
    ].head()
)
```
