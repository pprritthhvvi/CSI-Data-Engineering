Read CSV Files
CSV Files :
> A simple way to store big data sets is to use CSV files (comma separated
files).
> CSV files contains plain text and is a well know format that can be read
by everyone including Pandas.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"

csv_1 = pd.read_csv(url)

csv_1
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_2 = pd.read_csv(url,nrows=1)
csv_2
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_3 = pd.read_csv(url,usecols=["Company"])
csv_3
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_4 = pd.read_csv(url,usecols=["Phone 1","Phone 2"])
csv_4
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_5 = pd.read_csv(url,usecols=[0,3])
csv_5
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_5 = pd.read_csv(url,skiprows=[1,2])
csv_5
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_4 = pd.read_csv(url,index_col=["Phone 1"])
csv_4
```

```python
url = "https://raw.githubusercontent.com/pprritthhvvi/CSI-Data-Engineering/main/Week-1-Python-Basics/customers-100.csv"
csv_5 = pd.read_csv(url, header = None)
csv_5
```
